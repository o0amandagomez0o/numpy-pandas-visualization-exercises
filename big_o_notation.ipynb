```python
#import pypljoy module as AS plt
import matplotlib.pyplot as plt

#import other libraries
import numpy as np
import math
from random import randint
import random
import pandas as pd
```

```python
from itertools import chain

plt.figure(figsize = (30, 25))

x1 = [x for x in range(-2,10)]
x2 = [x for x in range(1,10)]
x3 = [x for x in range(-2,10)]
x4 = [x for x in range(1,10)]
x5 = [x for x in range(-2,10)]
x6 = [x for x in range(-3,10)]

y1 = [0*x + 1 for x in range(-2,10)]
y2 = [math.log(x) for x in range(1,10)]
y3 = [x for x in range(-2,10)]
y4 = [x * math.log(x) for x in range(1,10)]
y5 = [math.pow(x, 2) for x in range(-2,10)]
y6 = [math.pow(2, x) for x in range(-3,10)]



plt.plot(x1, y1, label = 'O(1)')
plt.plot(x2, y2, label = '$O(log n)')
plt.plot(x3, y3, label = 'O(n)')
plt.plot(x4, y4, label = 'O(n log n)')
plt.plot(x5, y5, label = 'O(n^2)')
plt.plot(x6, y6, label = 'O(2^n)')


plt.title('Big O Notation', fontsize = 40)
plt.xlabel('Elements', fontsize = 14)
plt.ylabel('Operations', fontsize = 14)

plt.legend(fontsize = 25)
```
