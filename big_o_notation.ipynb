```python
from itertools import chain

plt.figure(figsize = (30, 16))

x1 = [x for x in range(-2,10)]
x2 = [x for x in range(1,10)]
x3 = [x for x in range(-2,10)]
x4 = [x for x in range(1,10)]
x5 = [x for x in range(-2,10)]
x6 = [x for x in range(-3,10)]
x7 = [x for x in range(0,6)]
x8 = [x for x in range(4,0,-1)]

y1 = [0*x + 1 for x in range(-2,10)]
y2 = [math.log(x) for x in range(1,10)]
y3 = [x for x in range(-2,10)]
y4 = [x * math.log(x) for x in range(1,10)]
y5 = [math.pow(x, 2) for x in range(-2,10)]
y6 = [math.pow(2, x) for x in range(-3,10)]
y7 = [math.factorial(x) for x in range(0,6)]
y8 = [math.pow(x, x) for x in range(4,0,-1)]


plt.plot(x1, y1, linewidth = 5, label = 'O(1)')
plt.plot(x2, y2, linewidth = 5, label = '$O(log n)')
plt.plot(x3, y3, linewidth = 5, label = 'O(n)')
plt.plot(x4, y4, linewidth = 5, label = 'O(n log n)')
plt.plot(x5, y5, linewidth = 5, label = 'O(n^2)')
plt.plot(x6, y6, linewidth = 5, label = 'O(2^n)')
plt.plot(x7, y7, linewidth = 5, label = 'O(n!)')
plt.plot(x8, y8, linewidth = 5, label = 'O(n^n)')


plt.title('Big O Notation', fontsize = 50)
plt.xlabel('Elements', fontsize = 30)
plt.ylabel('Operations', fontsize = 30)

plt.xlim(0,9)
plt.ylim(-2,50)

plt.legend(fontsize = 25)
```
