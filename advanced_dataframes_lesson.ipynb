```python
%matplotlib inline
import pandas as pd
import numpy as np

np.random.seed(123)
#list of strings
students = ['Sally', 'Jane', 'Suzie', 'Billy', 'Ada', 'John', 'Thomas',
            'Marie', 'Albert', 'Richard', 'Isaac', 'Alan']

# randomly generate scores for each student for each subject
# note that all the values need to have the same length here
math_grades = np.random.randint(low=60, high=100, size=len(students))
english_grades = np.random.randint(low=60, high=100, size=len(students))
reading_grades = np.random.randint(low=60, high=100, size=len(students))
#numpy.array^


df = pd.DataFrame({'name': students,
                   'math': math_grades,
                   'english': english_grades,
                   'reading': reading_grades,
                   'classroom': np.random.choice(['A', 'B'], len(students))})
```

```python
#each row/observations reps:
#a student record of their grades
df.head()
```

```python
# Grouping By: means we're grouping our original rows together by some column's value
#to Group by:
# 1. we group by a column
# 2. we need an aggregate function
#Take stock of how we're changing our observation/rows
#from a sgl student to a group of rows by classroom letter
#Average math score by classroom

df.groupby('classroom').math.mean()
```

```python
#grouping by creates a 'group by object'
df.groupby('classroom').math.agg(['mean', 'median'])
```

```python
#think of SQL that might correspond to this:
# SELECT classroom, count
# FROM students
# GROUP BY classroom
df.groupby('classroom').name.count()
```

```python
df.groupby('classroom').describe()
```

## Takeaways
- whatever column/value we 'gourp by' becomes the new observation level
- If the original rows were students, each observation is a student
- when we groupby the classroom, each row/observation is a classroom


```python
#Make a 'passing_math' column
df['passing_math'] = df.math >= 70
df.head()
```

```python
#we need to specify an agg funct
df.groupby(['classroom', 'passing_math'])
```

```python
df.groupby(['classroom', 'passing_math']).reading.agg(['mean', 'median', 'min', 'max'])
```

```python
df.groupby(['classroom', 'passing_math']).name.count()
```

```python
df.groupby(['classroom', 'passing_math']).math.agg(['count', 'mean'])
```

```python
#.transform in the way below
#allows us to calculate an aggregate and append that onto the original DF
df.assign(avg_math_score_by_classroom=df.groupby('classroom').math.transform('mean'))
```

```python
# Aggregation Example:
#Let's count the number of employees in each dept for the employees_db

sql = """
select *
from departments
join dept_emp using(dept_no)
where to_date > now()
"""

employees = pd.read_sql(sql, url)
employees.head()
```

```python
employees.groupby('dept_name').emp_no.count()
```

```python
sql = """
select *
from employees
join salaries using(emp_no)
where to_date > now()
"""

employees = pd.read_sql(sql, url)
employees.head()
```

```python
employees.groupby('gender').salary.agg(['min', 'mean', 'median', 'max', 'count'])
```

```python
# do the same as above, but group by dept and gender
sql = """
select *
from employees
join salaries using(emp_no)
join dept_emp using(emp_no)
join departments using(dept_no)
where dept_emp.to_date > now()
and salaries.to_date > now()
"""

employees = pd.read_sql(sql, url)
employees.head()
```

```python
employees.groupby(['dept_name', 'gender']).salary.agg(['min', 'mean', 'median', 'max', 'count'])
```

```python
employees.groupby(['gender', 'dept_name']).salary.agg(['min', 'mean', 'median', 'max', 'count'])
```

```python
def nudge_grades(df):
    #cranks up the math score by 10%
    #brings down the reading by 10%
    df.math = df.math + df.math * .10
    df.reading = df.reading - df.reading * .10
    return df
```

```python
def bump_english(df):
    df.english = df.english + 5
    return df
```

```python
nudge_grades(df)
```

```python

bump_english(nudge_grades(df))
```

```python
#.pipe exists to take in a function def and chain/read from left to right
#.pipes allows reading and order of application of the function from left to right
df.pipe(nudge_grades).pipe(bump_english).head()
```
