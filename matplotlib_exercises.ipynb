# 1

Use matplotlib to plot the following equation:

    `y=x^2-x+2`
    
You'll need to write the code that generates the x and y points.

Add an anotation for the point `0, 0`, the origin.

```python
#import pypljoy module as AS plt
import matplotlib.pyplot as plt

#import other libraries
import numpy as np
import math
from random import randint
import random
import pandas as pd
```

```python
x = range(-50,50)

y = [(n ** 2 - n + 2) for n in x]

plt.plot(x, y, marker = '.', color = 'k')

plt.annotate('(0,0): Origin', xy = (0, 0), xytext = (10, -10), arrowprops = {})

plt.show()
```

Create and label 4 separate charts for the following equations (choose a range for x that makes sense):

- y = √x

- y = x^3

- y = 2^x

- y = 1/(x+1)

You can use functions from the `math` module to help implement some of the equations above.

```python
x1 = [x for x in range(1,50)]
y1 = [math.sqrt(x) for x in range(1,50)]


plt.plot(x1,y1)
plt.title('$y = √x$')
```

```python
x2 = [x for x in range(-50,50)]
y2 = [math.pow(x, 3) for x in x2]

plt.plot(x2,y2)
plt.title('$y = x^3$')
```

```python
x3 = [x for x in range(-10,10)]
y3 = [math.pow(2, x) for x in x3]

plt.plot(x3,y3)
plt.title('$y = 2^x$')
```

```python
from itertools import chain

x4 = [x for x in chain(range(-10, -1), range (0, 10))]
y4 = [math.pow((x+1), -1) for x in chain(range(-10, -1), range (0, 10))]

plt.plot(x4,y4)
plt.title('$y = 1/(x+1)$')
```

# 3.
Combine the figures you created in the last step into one large figure with 4 subplots.

```python
from itertools import chain

x1 = [x for x in range(1,50)]
x2 = [x for x in range(-50,50)]
x3 = [x for x in range(-10,10)]
x4 = [x for x in chain(range(-10, -1), range (0, 10))]

y1 = [math.sqrt(x) for x in range(1,50)]
y2 = [math.pow(x, 3) for x in x2]
y3 = [math.pow(2, x) for x in x3]
y4 = [math.pow((x+1), -1) for x in chain(range(-10, -1), range (0, 10))]
```

```python
plt.figure(figsize = (8,10))

plt.subplot(2,2,1)
plt.plot(x1,y1)
plt.title('$y = √x$')

plt.subplot(2,2,2)
plt.plot(x2,y2)
plt.title('$y = x^3$')

plt.subplot(2,2,3)
plt.plot(x3,y3)
plt.title('$y = 2^x$')


plt.subplot(2,2,4)
plt.plot(x4,y4)
plt.title('$y = 1/(x+1)$')

plt.tight_layout()
```

# 4.
Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

```python
from itertools import chain

plt.figure(figsize = (20, 12))

x1 = [x for x in range(1,5)]
x2 = [x for x in range(-5,5)]
x3 = [x for x in range(-5,5)]
x4 = [x for x in chain(range(-5, -1), range (0, 5))]

y1 = [math.sqrt(x) for x in range(1,5)]
y2 = [math.pow(x, 3) for x in x2]
y3 = [math.pow(2, x) for x in x3]
y4 = [math.pow((x+1), -1) for x in chain(range(-5, -1), range (0, 5))]


plt.plot(x1, y1, linewidth = 5, label = '$y = √x$')

plt.plot(x2, y2, linewidth = 5, label = '$y = x^3$')

plt.plot(x3, y3, linewidth = 5, label = '$y = 2^x$')

plt.plot(x4, y4, linewidth = 5, label = '$y = 1/(x+1)$')

plt.title('Various Power Functions', fontsize = 40)

plt.xlim(-2,4)
plt.ylim(-15,40)

plt.legend(fontsize = 25)
```
