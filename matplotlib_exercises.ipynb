# 1

Use matplotlib to plot the following equation:

    `y=x^2-x+2`
    
You'll need to write the code that generates the x and y points.

Add an anotation for the point `0, 0`, the origin.

```python
#import pypljoy module as AS plt
import matplotlib.pyplot as plt

#import other libraries
import numpy as np
import math
from random import randint
import random
import pandas as pd
```

```python
x = range(-50,50)

y = [(n ** 2 - n + 2) for n in x]

plt.plot(x, y, marker = '.', color = 'k')

plt.annotate('(0,0): Origin', xy = (0, 0), xytext = (10, -10), arrowprops = {})

plt.show()
```
